## Fake News Classification using Machine Learning and Natural Language Processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

#### Importing training dataset

```python
df = pd.read_csv("train_data.csv")
df.head()
```

#### Getting independent and dependent Features

```python
# dependent feature is the label 
# label = 1 -> unreliable news
# label = 0 -> reliable news 
X = df.drop('label',axis=1) # getting all the independent features
X.head()
```

```python
# dependent feature/target
y = df['label']
y.head()
```

```python
df.shape
```

```python
# dropping NaN values
df = df.dropna()
df.head()
```

#### Text Preprocessing

```python
# FOR IMPLEMENTING BAG OF WORDS
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
# creating a copy of dataframe
msg = df.copy()
# to counter the affect of dropping NaNs for proper indexing
msg.reset_index(inplace=True)
msg.head(15)
```

```python
# Stemming and Removing all the stop words(unimportant words for analysis)
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(len(msg)):
    s = re.sub('[^a-zA-Z]',' ',msg['title'][i])
    s = s.lower()
    s = s.split()
    s = [ps.stem(w) for w in s if not w in stopwords.words('english')]
    s = ' '.join(s)
    corpus.append(s)
corpus
```

#### (I) Using Bag of Words Model

```python
# Applying Count Vectorizer to create a Bag of Words Model
cv = CountVectorizer(max_features = 5000,ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
X.shape
```

```python
y = msg['label']
```

```python
# Splitting test and train dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
cv.get_feature_names()[:20]
```

```python
cv.get_params()
```

```python
x_df = pd.DataFrame(X_train,columns=cv.get_feature_names())
x_df.head(15)
```

```python
# PLOTTING CONFUSION MATRIX FOR ANALYSIS OF ACCURACY
def plot_confusion_matrix(cm,classes,normalize=False,title="Confusion Matrix",cmap=plt.cm.Blues):
    plt.imshow(cm,interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)
    if (normalize):
        cm = cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
        print("Normalized Confusion matrix")
    else:
        print("Confusion matrix,No Normalization")
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
```

#### MultinomialNB Algorithm for News Classification

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
nb = MultinomialNB()
```

```python
nb.fit(X_train,y_train)
y_predicted = nb.predict(X_test)
score = accuracy_score(y_test,y_predicted)
cmat = confusion_matrix(y_test,y_predicted)
```

```python
print("ACCURACY SCORE : \n",score)
plot_confusion_matrix(cmat,classes=['FAKE NEWS','REAL NEWS'])
```

#### Passive Aggressive Classifier Algorithm

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.get_params()
```

```python
pac.fit(X_train,y_train)
y_pred_pac = pac.predict(X_test)
pac_score = accuracy_score(y_test,y_pred_pac)
print("ACCURACY SCORE : \n",pac_score)
pac_cmat = confusion_matrix(y_test,y_pred_pac)
plot_confusion_matrix(pac_cmat,classes = ['FAKE NEWS','REAL NEWS'])
```

#### Multinomial Classifier with Hyperparameter

```python
nbclassifier = MultinomialNB(alpha=0.1)
```

```python
# To find the best alpha parameter
prev_score = 0
for al in np.arange(0,1,0.1):
    sub_nb = MultinomialNB(alpha=al)
    sub_nb.fit(X_train,y_train)
    y_pred = sub_nb.predict(X_test)
    cur_score = accuracy_score(y_test,y_pred)
    if (cur_score > prev_score):
        nb_classifier = sub_nb
    print("Alpha: {}, Score:{}".format(al,cur_score))
```
